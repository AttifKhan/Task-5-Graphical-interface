```python
import tkinter as tk

# Create a Tkinter window
window = tk.Tk()
window.title("Menu Example")

# Function to handle the "File > Open" menu option
def open_file():
    # Placeholder function
    print("Open file")

# Function to handle the "File > Save" menu option
def save_file():
    # Placeholder function
    print("Save file")

# Function to handle the "Edit > Cut" menu option
def cut_text():
    # Placeholder function
    print("Cut text")

# Function to handle the "Edit > Copy" menu option
def copy_text():
    # Placeholder function
    print("Copy text")

# Function to handle the "Edit > Paste" menu option
def paste_text():
    # Placeholder function
    print("Paste text")

# Function to handle the "Help > About" menu option
def show_about():
    # Placeholder function
    print("About")

# Create a menu bar
menu_bar = tk.Menu(window)

# Create a "File" menu
file_menu = tk.Menu(menu_bar, tearoff=0)
file_menu.add_command(label="Open", command=open_file)
file_menu.add_command(label="Save", command=save_file)
file_menu.add_separator()
file_menu.add_command(label="Exit", command=window.destroy)

# Create an "Edit" menu
edit_menu = tk.Menu(menu_bar, tearoff=0)
edit_menu.add_command(label="Cut", command=cut_text)
edit_menu.add_command(label="Copy", command=copy_text)
edit_menu.add_command(label="Paste", command=paste_text)

# Create a "View" menu
view_menu = tk.Menu(menu_bar, tearoff=0)
view_menu.add_command(label="Toggle Fullscreen")
view_menu.add_command(label="Zoom In")
view_menu.add_command(label="Zoom Out")

# Create a "Help" menu
help_menu = tk.Menu(menu_bar, tearoff=0)
help_menu.add_command(label="About", command=show_about)

# Add the submenus to the main menu bar
menu_bar.add_cascade(label="File", menu=file_menu)
menu_bar.add_cascade(label="Edit", menu=edit_menu)
menu_bar.add_cascade(label="View", menu=view_menu)
menu_bar.add_cascade(label="Help", menu=help_menu)

# Configure the window to use the menu bar
window.config(menu=menu_bar)

# Run the Tkinter event loop
window.mainloop()
```
